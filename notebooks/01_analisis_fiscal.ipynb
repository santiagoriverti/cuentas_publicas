## 1. Resultado Primario y Financiero â€” Serie 2020-2026

El **resultado primario** es ingresos menos gastos primarios (excluye intereses de deuda).  
El **resultado financiero** incluye el pago de intereses.

```python
# Filtrar: mensual, total_adm_nacional
mask = (df_aif['periodo'] == 'mensual') & (df_aif['subsector'] == 'total_adm_nacional')
df_m = df_aif[mask].copy()

primario = df_m[df_m['concepto_codigo'] == 'XIV_RESULTADO_PRIMARIO'].set_index('fecha')['valor_millones_pesos']
financiero = df_m[df_m['concepto_codigo'] == 'XV_RESULTADO_FINANCIERO'].set_index('fecha')['valor_millones_pesos']

fig, ax = plt.subplots(figsize=(16, 6))
ax.bar(primario.index, primario.values / 1000, width=20,
       color=['#2ecc71' if v >= 0 else '#e74c3c' for v in primario.values],
       alpha=0.8, label='Resultado Primario')
ax.plot(financiero.index, financiero.values / 1000, 'o-', color='#2c3e50',
        linewidth=2, markersize=4, label='Resultado Financiero')
ax.axhline(0, color='black', linewidth=0.8)
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x:,.0f}'))
ax.set_title('Resultado Primario y Financiero â€” Adm. Nacional mensual (miles de MM ARS)', fontsize=14)
ax.set_ylabel('Miles de millones de pesos')
ax.legend()
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Tabla resumen anual
df_annual = df_m[df_m['concepto_codigo'].isin(['XIV_RESULTADO_PRIMARIO', 'XV_RESULTADO_FINANCIERO'])].copy()
df_annual['anio'] = df_annual['fecha'].dt.year
pivot = df_annual.groupby(['anio', 'concepto_codigo'])['valor_millones_pesos'].sum().unstack() / 1_000_000
pivot.columns = ['Resultado Financiero (MM MM ARS)', 'Resultado Primario (MM MM ARS)']
print('\nResumen Anual (billones de ARS):')
print(pivot.round(2).to_string())
```

## 2. ComposiciÃ³n del Gasto â€” Ajuste Milei 2024-2026

```python
# Gastos corrientes vs capital
gastos_cols = {
    'II_GASTOS_CORRIENTES': 'Gastos Corrientes',
    'V_GASTOS_CAPITAL': 'Gastos de Capital',
}
mask2 = (df_aif['periodo'] == 'mensual') & (df_aif['subsector'] == 'total_adm_acional') 
# Nota: usar total_general si disponible para 2026
mask2 = (
    (df_aif['periodo'] == 'mensual') &
    (df_aif['subsector'].isin(['total_adm_nacional', 'total_general'])) &
    (df_aif['concepto_codigo'].isin(gastos_cols.keys()))
)

df_gastos = df_aif[mask2].copy()
# Priorizar total_general sobre total_adm_nacional cuando existen ambos
df_gastos = df_gastos.sort_values('subsector').drop_duplicates(['fecha', 'concepto_codigo'], keep='last')

pivot_g = df_gastos.pivot_table(index='fecha', columns='concepto_codigo', values='valor_millones_pesos')
pivot_g.columns = [gastos_cols.get(c, c) for c in pivot_g.columns]

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

for ax, col in zip(axes, pivot_g.columns):
    series = pivot_g[col].dropna()
    ax.bar(series.index, series.values / 1000, width=20, color='#3498db', alpha=0.7)
    ax.set_title(col)
    ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x:,.0f}'))
    ax.set_ylabel('Miles de MM ARS')
    ax.grid(axis='y', alpha=0.3)

plt.suptitle('EvoluciÃ³n del Gasto PÃºblico â€” Adm. Nacional mensual', fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
```

## 3. Transferencias a Provincias â€” CoparticipaciÃ³n y Discrecionalidad

```python
transf_prov_corrientes = df_aif[
    (df_aif['concepto_codigo'] == 'II4b1_TRANSF_PROVINCIAS_CABA') &
    (df_aif['periodo'] == 'mensual') &
    (df_aif['subsector'] == 'tesoro_nacional')
].set_index('fecha')['valor_millones_pesos']

transf_prov_capital = df_aif[
    (df_aif['concepto_codigo'] == 'V2a_TRANSF_CAPITAL_PROVINCIAS') &
    (df_aif['periodo'] == 'mensual') &
    (df_aif['subsector'] == 'tesoro_nacional')
].set_index('fecha')['valor_millones_pesos']

fig, ax = plt.subplots(figsize=(16, 5))
ax.bar(transf_prov_corrientes.index, transf_prov_corrientes.values / 1000,
       width=20, label='Transf. Corrientes a Provincias/CABA', color='#9b59b6', alpha=0.8)
ax.bar(transf_prov_capital.index, transf_prov_capital.values / 1000,
       width=20, bottom=transf_prov_corrientes.reindex(transf_prov_capital.index).fillna(0) / 1000,
       label='Transf. Capital a Provincias/CABA', color='#e67e22', alpha=0.8)

ax.set_title('Transferencias del Tesoro a Provincias y CABA (mensual)', fontsize=13)
ax.set_ylabel('Miles de MM ARS')
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x:,.0f}'))
ax.legend()
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Tabla anual
df_prov = pd.DataFrame({
    'Transf. Corrientes': transf_prov_corrientes,
    'Transf. Capital': transf_prov_capital
})
df_prov.index = pd.to_datetime(df_prov.index)
print('Transferencias anuales a Provincias/CABA (Tesoro, miles MM ARS):')
print((df_prov.groupby(df_prov.index.year).sum() / 1000).round(1).to_string())
```

## 4. Ingresos: ComposiciÃ³n Tributaria

```python
mask_ing = (
    (df_aif['periodo'] == 'mensual') &
    (df_aif['subsector'] == 'total_adm_nacional') &
    (df_aif['concepto_codigo'].isin(['I_INGRESOS_CORRIENTES', 'I1_INGRESOS_TRIBUTARIOS',
                                      'I2_APORTES_SEG_SOCIAL']))
)

pivot_ing = df_aif[mask_ing].pivot_table(
    index='fecha', columns='concepto_codigo', values='valor_millones_pesos'
)

fig, ax = plt.subplots(figsize=(16, 5))
if 'I1_INGRESOS_TRIBUTARIOS' in pivot_ing.columns:
    ax.fill_between(pivot_ing.index, pivot_ing['I1_INGRESOS_TRIBUTARIOS'] / 1000,
                    alpha=0.7, color='#27ae60', label='Ingresos Tributarios')
if 'I2_APORTES_SEG_SOCIAL' in pivot_ing.columns:
    ax.fill_between(pivot_ing.index, pivot_ing['I2_APORTES_SEG_SOCIAL'] / 1000,
                    alpha=0.6, color='#2980b9', label='Aportes Seg. Social')
if 'I_INGRESOS_CORRIENTES' in pivot_ing.columns:
    ax.plot(pivot_ing.index, pivot_ing['I_INGRESOS_CORRIENTES'] / 1000,
            'k-', linewidth=1.5, label='Total Ingresos Corrientes')

ax.set_title('EvoluciÃ³n de Ingresos â€” Adm. Nacional mensual', fontsize=13)
ax.set_ylabel('Miles de MM ARS')
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x:,.0f}'))
ax.legend()
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

## 5. DesagregaciÃ³n por Subsector Institucional

```python
# Resultado primario por subsector
mask_subsect = (
    (df_aif['concepto_codigo'] == 'XIV_RESULTADO_PRIMARIO') &
    (df_aif['periodo'] == 'mensual') &
    (df_aif['subsector'].isin(['tesoro_nacional', 'rec_afectados',
                                'org_descentralizados', 'inst_seg_social']))
)

pivot_ss = df_aif[mask_subsect].pivot_table(
    index='fecha', columns='subsector', values='valor_millones_pesos'
)

labels = {
    'tesoro_nacional': 'Tesoro Nacional',
    'rec_afectados': 'Rec. Afectados',
    'org_descentralizados': 'Org. Descentralizados',
    'inst_seg_social': 'Seg. Social (ANSES)',
}

colors = ['#e74c3c', '#f39c12', '#3498db', '#2ecc71']

fig, ax = plt.subplots(figsize=(16, 6))
for col, color in zip(pivot_ss.columns, colors):
    ax.plot(pivot_ss.index, pivot_ss[col] / 1000, label=labels.get(col, col),
            linewidth=1.5, color=color)

ax.axhline(0, color='black', linewidth=0.8, linestyle='--')
ax.set_title('Resultado Primario por Subsector Institucional (mensual)', fontsize=13)
ax.set_ylabel('Miles de MM ARS')
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{x:,.0f}'))
ax.legend()
ax.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
